## Data from original paper

```python
all_stocks_5yr = pd.read_csv('stocks_data/all_stocks_5yr.csv').head()
all_stocks_5yr.info()
```

```python
import matplotlib.pyplot as plt
df = all_stocks_5yr
# Convert 'date' column to datetime type
df['date'] = pd.to_datetime(df['date'])

# Display the first few rows of the dataframe
print(df.head())

# Basic summary statistics
print(df.describe())

# Check for missing values
print(f'Missing values: \n{df.isnull().sum()}')

# Data visualization

# Line plot of 'close' prices
plt.plot(df['date'], df['close'])
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.title('Closing Prices over Time')
plt.show()

# Box plot of 'volume' column
plt.boxplot(df['volume'])
plt.xlabel('Volume')
plt.title('Volume Distribution')
plt.show()

# Scatter plot of 'high' vs 'low'
plt.scatter(df['high'], df['low'])
plt.xlabel('High Price')
plt.ylabel('Low Price')
plt.title('High vs Low Prices')
plt.show()

# Histogram of 'open' prices
plt.hist(df['open'], bins=10)
plt.xlabel('Open Price')
plt.ylabel('Frequency')
plt.title('Distribution of Open Prices')
plt.show()
```

```python
import seaborn as sns
data = all_stocks_5yr
data.hist(figsize=(10, 6))
plt.tight_layout()
plt.show()

correlation = data.corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()

sns.scatterplot(data=data, x='open', y='close', hue='Name')
plt.title('Open vs. Close Prices')
plt.show()

data['date'] = pd.to_datetime(data['date'])
data.set_index('date', inplace=True)
data[['open', 'high', 'low', 'close']].plot(figsize=(12, 6))
plt.title('Stock Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```
